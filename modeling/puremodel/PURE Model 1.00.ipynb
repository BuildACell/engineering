# Mass Action Kinetic Model of the PURExpressÂ® System
##ODEs to Model are:
mRNA' = ktx - km[mRNA]
protein' = ktl[mRNA] - kp[protein]

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import scipy as sp
import math

ktx = 1.0
km = 1.0
ktl = 1.0
kp = 1.0

def massActionModel(y, t):
    mRNA = y[0]
    protein = y[1]
    return [(ktx - km*mRNA), (ktl*mRNA - kp*protein)]

plt.style.use("ggplot") # GGplot is awesome.
plt.rcParams['figure.figsize'] = (10.0, 8.0)

t = np.arange (0, 10, 0.01)
y = []
y = odeint(massActionModel, (0,0), t)

plt.plot(t,y[:, 1])

plt.xlabel("Time")
plt.ylabel("Protein Concentration")

plt.show()

```
